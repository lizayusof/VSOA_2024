```python
#import modules and libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#put the path for the file
df = pd.read_csv('20p0msun_data/LOGS/profile1.data',skiprows=[0,1,2,3,4],sep='\s+')
```

```python
params = {'backend': 'TKAgg',
'axes.labelsize':  16,
'legend.fontsize': 12,
'xtick.labelsize': 14,
'ytick.labelsize': 16,
'figure.figsize': (8,8),
'text.usetex':True,
'figure.dpi': 90}

plt.rcParams.update(params)
```

```python
#assign the variables
mass = df['mass']
radius = df['logR']
T = 10**df['logT']
L = df['luminosity']
Rho = df['logRho']
H = df['h1']
He3 = df['he3']
He4 = df['he4']
C12 = df['c12']
N14 = df['n14']
```

```python
#start plotting routine
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(radius, mass)
ax[0, 0].set_xlabel('Radius, $r$')
ax[0, 0].set_ylabel('Mass, M$_\odot$')

ax[0, 1].plot(radius, L)
ax[0,1].invert_xaxis()
ax[0, 1].set_xlabel('Radius, $r$')
ax[0, 1].set_ylabel('Luminosity')


ax[1, 0].plot(radius,T)
ax[1, 0].set_xlabel('Radius, $r$')
ax[1, 0].set_ylabel('Temperature')


ax[1, 1].plot(radius, Rho)
ax[1, 1].set_xlabel('Radius, $r$')
ax[1, 1].set_ylabel('Density, $\\rho$')
plt.tight_layout()
```
